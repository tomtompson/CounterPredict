```python
import pandas as pd
```

```python
teams = pd.read_csv("../teams/top_50_teams.csv")
matches = pd.read_csv("../matches/top_50_team_matches.csv")
failed_player_stats = pd.read_csv("../matches/failed_match_player_stats.csv")
player_stats = pd.read_csv("../matches/top_50_team_match_player_stats.csv")
```

```python
print(len(matches))
failed_indexes = matches[matches["match_id"].isin(failed_player_stats["match_id"])].index

matches = matches.drop(failed_indexes)
print(len(matches))
```

```python

print(teams.describe())

print(matches.describe())

print(player_stats.describe())
```

```python
empty_stats_teams = teams[teams.isna().any(axis=1)]
empty_stats_teams
```

```python
empty_stats_match = matches[matches.isna().any(axis=1)]
empty_stats_match
```

```python
empty_stats = player_stats[player_stats.isna().any(axis=1)]
empty_stats
```

```python
print(f"len matches {len(matches)}")
set_missed_matches = set(empty_stats["match_id"])

matches.drop(matches[matches.match_id.isin(set_missed_matches)].index, axis = 0, inplace = True)
print(f"len matches after filter {len(matches)}")
```

```python
team_features = [
    "id",
    "hltv_placement",
    "hltv_points",
    "valve_placement",
    "valve_points",
]

team_1_org = teams[team_features].rename(columns={
    "id": "team_1_id",
    "hltv_placement": "team_1_hltv_placement",
    "hltv_points": "team_1_hltv_points",
    "valve_placement": "team_1_valve_placement",
    "valve_points": "team_1_valve_points",
})

team_2_org = teams[team_features].rename(columns={
    "id": "team_2_id",
    "hltv_placement": "team_2_hltv_placement",
    "hltv_points": "team_2_hltv_points",
    "valve_placement": "team_2_valve_placement",
    "valve_points": "team_2_valve_points",
})

match_df = matches.merge(team_1_org, on="team_1_id", how="left")
match_df = match_df.merge(team_2_org, on="team_2_id", how="left")

match_df
```

```python
player_stats["team_side"] = player_stats.apply(
    lambda row: "team_1" if row["player_team_id"] == row["team_1_id"] else "team_2",
    axis=1
)

player_stats
```

```python
player_stats["player_slot"] = (
    player_stats
    .groupby(["match_id", "team_side"])
    .cumcount() + 1
)

player_feature_cols = [
    "player_name",
    "player_id",
    "rating_3_0",
    "rating_3_0_t_rating",
    "rating_3_0_ct_rating",
    "round_swing",
    "dpr",
    "kast",
    "multi_kill",
    "adr",
    "kpr",
    "firepower_combined_score",
    "entrying_combined_score",
    "trading_combined_score",
    "opening_combined_score",
    "clutching_combined_score",
    "sniping_combined_score",
    "utility_combined_score",
]

player_wide = player_stats[
    ["match_id", "team_side", "player_slot"] + player_feature_cols
].pivot(
    index="match_id",
    columns=["team_side", "player_slot"],
    values=player_feature_cols
)

player_wide.columns = [
    f"{team_side}_player_{player_slot}_{feature}"
    for feature, team_side, player_slot in player_wide.columns
]

player_wide = player_wide.reset_index()

final_df = match_df.merge(player_wide, on="match_id", how="left")

final_df["target_team_1_win"] = final_df["team_1_winner"].astype(int)
```

```python
final_df["winner_team_id"] = final_df.apply(
    lambda row: row["team_1_id"] if row["team_1_winner"] else row["team_2_id"],
    axis=1
)

final_df["winner_team_name"] = final_df.apply(
    lambda row: row["team_1"] if row["team_1_winner"] else row["team_2"],
    axis=1
)

final_df
```

```python
empty_stats = final_df[final_df.isna().any(axis=1)]
empty_stats
```

```python
final_df.to_csv('training_dataset.csv', index=False)
```
